# chimpanzee

```python
adata_chimpanzee = sc.read_h5ad('./dataset/raw-testis-chimpanzee-Murat.h5ad')
adata_chimpanzee
```

```python
gene_array = pd.read_csv('./dataset/gene_matches_human2chimpanzee.csv').to_numpy()
gene_dict = {item[1]: item[0] for item in gene_array}
gene_sub = list(set(adata_chimpanzee.var.index) & set(gene_dict.keys()))
```

```python
adata_chimpanzee = adata_chimpanzee[:, gene_sub].copy()
adata_chimpanzee
```

```python
adata_chimpanzee.var.index = adata_chimpanzee.var.loc[:, 'gene_symbol'].apply(lambda x: gene_dict[x])
adata_chimpanzee
```

```python
adata_scvi = adata_11species.copy()
```

# pyliger

```python
import pyliger
import numpy as np
```

```python
adata_pyliger = adata_11species.copy()
adata_pyliger.obs_names_make_unique()
adata_pyliger.var_names_make_unique()
adata_pyliger
```

```python
adata_pyliger.X.max(), adata_pyliger.layers['counts'].max()
```

```python
adata_pyliger.obs.index.name = 'cell'
adata_pyliger.var.index.name = 'gene'
```

```python
batches = adata_pyliger.obs.loc[:, 'batch'].unique()
batches
```

```python
adata_list = []
for batch in batches:
    temp = adata_pyliger[adata_pyliger.obs.loc[:, 'batch'].copy() == batch]
    temp.uns['sample_name'] = batch
    temp.obs.index = [batch + '_' + str(item) for item in list(range(len(temp)))]
    temp.obs.index.name = 'cell'
    temp.var.index.name = 'gene'
    adata_list.append(temp)
adata_list
```

```python
species_liger = pyliger.create_liger(adata_list)
pyliger.normalize(species_liger)
pyliger.select_genes(species_liger)
pyliger.scale_not_center(species_liger)
pyliger.optimize_ALS(species_liger, k = 20)
pyliger.quantile_norm(species_liger)
```

```python
h_norm = np.vstack([adata.obsm['H_norm'] for adata in species_liger.adata_list])
adata_pyliger.obsm['X_pyliger'] = h_norm
```

```python
sc.pp.neighbors(adata_pyliger, use_rep='X_pyliger')
sc.tl.umap(adata_pyliger)
sc.pl.umap(adata_pyliger, color=['batch', 'cell_ontology_class'], wspace=0.2)
```
